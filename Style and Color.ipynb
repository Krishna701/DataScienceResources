# Style and Color

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
tips = sb.load_dataset('tips')
```

## Styles

You can set particular styles:

```python
sb.countplot(x='sex',data=tips)
```

```python
sb.set_style('dark')
sb.countplot(x='sex',data=tips)
```

```python
sb.set_style('ticks')
sb.countplot(x='sex',data=tips,palette='deep')
```

## Spine Removal

```python
sb.countplot(x='sex',data=tips)
sb.despine()
```

```python
sb.countplot(x='sex',data=tips)
sb.despine(left=True)
```

## Size and Aspect

You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sb.countplot(x='sex',data=tips)
```

```python
# Grid Type Plot
sb.lmplot(x='total_bill',y='tip',height=2,aspect=4,data=tips)
```

## Scale and Context

The set_context() allows you to override default parameters:

```python
sb.set_context("notebook",font_scale=) #paper, notebook, talk, poster
sb.countplot(x='sex',data=tips,palette='coolwarm')
```
