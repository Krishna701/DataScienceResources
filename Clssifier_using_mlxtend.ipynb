```python
from sklearn.svm import SVC # Scikit-learn -> Support Vector Machine -> Support Vector Classifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#pip install mlxtend
from mlxtend.plotting import plot_decision_regions
```

```python
#Data Generation -> Numpy
```

```python
X = np.array([[x/10,y/10] for x in range(20) for y in range(20)])
y = np.array([1 if (sum((x-np.array([1.5,1.5]))**2) < 0.5) else 0 for x in X])
```

```python
plt.axes()
plt.scatter(*zip(*X), c=y, cmap= 'binary')
```

```python
clf = LogisticRegression(solver='liblinear')
clf.fit(X,y)
plot_decision_regions(X, y, clf, legend=2)
plt.xlim([0,2])
plt.ylim([0,2])
print(confusion_matrix(y, clf.predict(X)))
```

```python
clf = SVC(gamma = 300)
y_ = y.copy()
y_[50]=1
y_[51]=1
y_[52]=1
y_[53]=1
clf.fit(X,y_)
plot_decision_regions(X, y_, clf, legend=2)
plt.xlim([0,2])
plt.ylim([0,2])
print(confusion_matrix(y_, clf.predict(X)))
```

```python
clf = KNeighborsClassifier()
y_ = y.copy()
y_[50]=1
y_[51]=1
y_[52]=1
y_[53]=1
clf.fit(X,y_)
plot_decision_regions(X, y_, clf, legend=2)
plt.xlim([0,2])
plt.ylim([0,2])
print(confusion_matrix(y_, clf.predict(X)))
```

```python
clf = DecisionTreeClassifier()
y_ = y.copy()
y_[50]=1
y_[51]=1
y_[52]=1
y_[53]=1
clf.fit(X,y_)
plot_decision_regions(X, y_, clf, legend=2)
plt.xlim([0,2])
plt.ylim([0,2])
print(confusion_matrix(y_, clf.predict(X)))
```
